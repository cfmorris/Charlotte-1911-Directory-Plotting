## 1. Project Overview

---
Goals: 
   1. to georeference addresses in bulk, resolving them to lat/long coordinates.
   2. to visually plot those coordinates inside Jupyter Noptebook.
   3. to achieve goals 1. and 2. with a process that can easily be reapplied by anyone to any dataset.
---

Our project begins assuming the availability of a dataset.  Our dataset was created by applying ocular character recognition (OCR) techniques to the 1911 Charlotte City Registry, which is available through the Internet Archives at https://archive.org/details/charlottenorthca1911pied. The data were then refined with OpenRefine by Dr. Marciano, who separated the data contained in each entry into useful categories such as name, profession, race, address, etc.  This project is an attempt to resolve those addresses into coordinate data so that each entry can be plotted on a map, allowing the people of 1911 Charlotte to be visualized individually and in aggregate.

# Computation Thinking

This project involved almost every practice identified by the Computational Thinking taxonomy above.  

1. Data Practices: Coordinated data was created after manipulating and refining the given dataset. This data was then visualized by plotting it on a map and those plots were then analyzed in two ways. First, to determine the accuracy of the data and second, to identify problem types that can be clustered and addressed in future iterations.

2. Modeling and Simulation: Before we attempted to georeference the entire 15,000-entry dataset provided by Dr. Marciano, we designed and used two models/limited simulations of what we intended to do with the Charlotte data. Having extracted lat/long data for a small number of entries, we first tested the accuracy of the georeferencer service (OpenStreetMaps via Nomitatum) by georeferencing a known quantity (the Eiffel Tower). Then, we identified and troubleshot issues arising from running a limited data set of ~150 entries through Nomitatum. This let us estimate what percent of entries could be reliably georeferenced, how to start improving the ratio, and whether or not we would be able to draw meaningful conclusions from the data once it was visualized. If we could georeference a comfortable percentage of the entries, we planned to map sort and map our entries with Folium 1) by race and 2) by a selection of job types.

3. Computational Problem Solving: Having chosen effective computational tools (geopandas, geopy, geopandas, etc.) we began the programming process by running our full batch through Nomitatum to convert address info into more useable lat/long coordinates. We took full advantage of J. Notebook's emphasis on developing modular solutions, and did a check on each section before proceeding, for example by writing a timeout parameter for Nomitatum that would take longer but ensure better results, which we then tested manually via running the Eiffel Tower's address. As we developed our final .csv file we troubleshot issues as we detected them, such as assigning a random value to entries with "null" values for their coordinates, in order to make mapping them easier. We applied the same concepts to each of our maps when the time came.

4. Systems Thinking: Once we mapped our data along the lines of race and profession, we started to consider what we saw from two perspectives, 1) the system of datapoints as a complex whole, and 2) the relationships between the datapoints, as well as the technical processes that rendered them. Doing so revealed a number of issues and limitations with the georeferencer that we had not anticipated. We came to the conclusion that the data was too inaccurate for us to use it to draw any interpretive conclusions about life in 1911 Charlotte, North Carolina.
